```python
# prepare independent and dependent variable
X = marketing.drop(['Response'], axis=1)
y = marketing['Response']
```

```python
# assign 'response' column to y 
y = marketing['Response'].astype('str')
# assign all columns except 'churned' to X
X = marketing.drop(['Response'], axis = 1)
```

```python
y.shape
```

```python
# check X dimension
X.shape
```

```python
#Create dummy variables for each cat. variable
cat_col = ['object']
cat_columns = list(X.select_dtypes(include=cat_col).columns)
cat_data = X[cat_columns]
cat_vars = cat_data.columns
```

```python
for var in cat_vars:
    cat_list = pd.get_dummies(X[var], prefix=var)
    X = X.join(cat_list)


data_vars = X.columns.values.tolist()
to_keep = [i for i in data_vars if i not in cat_vars]

#Create final dataframe
X_final = X[to_keep]
X_final.columns.values
```

```python
X_final = X_final.astype(float)
```

```python
# split data to train and test data
# with test size is 30% of overall data
# use stratified sampling
# and set RNG to 1000
X_train, X_test, y_train, y_test= train_test_split(
    X_final,
    y,
    test_size = 0.3,
    stratify = y,
    random_state = 1000
)
```

```python
classifier = LogisticRegression(max_iter=10000)
classifier.fit(X_train, y_train)
```
